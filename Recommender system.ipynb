## Using Pearson correlation b/w input Transaction and usersubset

```python
from math import sqrt
```

```python
pearson_dict={}

for Tid, group in subset2:
    group = group.sort_values(by='ItemCode') #lets sort the values first so that there's no confusion later
    inputid = dfi.sort_values(by='ItemCode')
    
    #the n value for the formula
    nrate = len(group)
    
    #lets put the common items b/w the i/p and the user subset in a temporary df
    temp_df = inputid[inputid['ItemCode'].isin(group['ItemCode'].tolist())]
    temp_rate = temp_df['NumberOfItemsPurchased'].values
    temp_grp = group['NumberOfItemsPurchased'].values
    
    #lets calculate the pearson correlation
    Sxx = sum([i**2 for i in temp_rate]) - pow(sum(temp_rate),2)/float(nrate)
    Syy = sum([i**2 for i in temp_grp]) - pow(sum(temp_grp),2)/float(nrate)
    Sxy = sum( i*j for i, j in zip(temp_rate, temp_grp)) - sum(temp_rate)*sum(temp_grp)/float(nrate)
    
    #If the denominator is different than zero, then divide, else, 0 correlation.
    if Sxx != 0 and Syy != 0:
        pearson_dict[Tid] = Sxy/sqrt(Sxx*Syy)
    else:
        pearson_dict[Tid] = 0
```

```python
pearson_dict.items()
```

### lets look at the similarityindex for these 100 Transactions

```python
pearsondf = pd.DataFrame.from_dict(pearson_dict, orient='index')
pearsondf.columns = ['similarityIndex']
pearsondf['TransactionId'] = pearsondf.index
pearsondf.index = range(len(pearsondf))
```

```python
pearsondf
```

#### top 50 Transactions with highest similarity index

```python
toptran=pearsondf.sort_values(by='similarityIndex', ascending=False)[0:50]
toptran.head()
```

#### merge the top 50 transactions with the rating df(rdf) to get the item codes and their rating(Numberofitemspurchased)

```python
toptranrate = toptran.merge(rdf, left_on='TransactionId', right_on='TransactionId', how='inner')
toptranrate.rename(columns = {'NumberOfItemsPurchased':'Rating'},inplace=True)
toptranrate.head()
```

#### lets calculate the weighted rating

```python
toptranrate['weightedrating'] = toptranrate['similarityIndex']* toptranrate['Rating']
toptranrate.head()
```

#### Applying a sum to the toptransactions after grouping it up by itemcode

```python
temptopTransactionrating = toptranrate.groupby('ItemCode').sum()[['similarityIndex','weightedrating']]
temptopTransactionrating.columns = ['sum_similarityIndex','sum_weightedRating']
temptopTransactionrating.head()
```

```python
#Creates an empty dataframe
recommendation_df = pd.DataFrame()
#Now we take the weighted average
recommendation_df['weighted average recommendation score'] = temptopTransactionrating['sum_weightedRating']/temptopTransactionrating['sum_similarityIndex']
recommendation_df['ItemCode'] = temptopTransactionrating.index
recommendation_df.head()
```

## Final function to generate frequently bought products

```python
#final function to generate frequently bought products
def predict(df,ip):
    rdf = df[['TransactionId','ItemCode','NumberOfItemsPurchased']]
    dfi = pd.DataFrame(ip)
    #select the transactions which have the i/p items in them
    usub = rdf[rdf['ItemCode'].isin(dfi['ItemCode'].values)]
    #sort the transactions by the length of the usersubset group where transactions with most nos. of i/p items at the top 
    subset = sorted(usub.groupby('TransactionId'),  key=lambda x: len(x[1]), reverse=True)
    
    
    pearson_dict={}

    for Tid, group in subset2:
        group = group.sort_values(by='ItemCode') #lets sort the values first so that there's no confusion later
        inputid = dfi.sort_values(by='ItemCode')
    
        #the n value for the formula
        nrate = len(group)
    
    #lets put the common items b/w the i/p and the user subset in a temporary df
        temp_df = inputid[inputid['ItemCode'].isin(group['ItemCode'].tolist())]
        temp_rate = temp_df['NumberOfItemsPurchased'].values
        temp_grp = group['NumberOfItemsPurchased'].values
    
    #lets calculate the pearson correlation
        Sxx = sum([i**2 for i in temp_rate]) - pow(sum(temp_rate),2)/float(nrate)
        Syy = sum([i**2 for i in temp_grp]) - pow(sum(temp_grp),2)/float(nrate)
        Sxy = sum( i*j for i, j in zip(temp_rate, temp_grp)) - sum(temp_rate)*sum(temp_grp)/float(nrate)
    
    #If the denominator is different than zero, then divide, else, 0 correlation.
        if Sxx != 0 and Syy != 0:
            pearson_dict[Tid] = Sxy/sqrt(Sxx*Syy)
        else:
            pearson_dict[Tid] = 0
    
    pearsondf = pd.DataFrame.from_dict(pearson_dict, orient='index')
    pearsondf.columns = ['similarityIndex']
    pearsondf['TransactionId'] = pearsondf.index
    pearsondf.index = range(len(pearsondf))
    toptran=pearsondf.sort_values(by='similarityIndex', ascending=False)[0:50]
    toptranrate = toptran.merge(rdf, left_on='TransactionId', right_on='TransactionId', how='inner')
    toptranrate.rename(columns = {'NumberOfItemsPurchased':'Rating'},inplace=True)
    
    toptranrate['weightedrating'] = toptranrate['similarityIndex']* toptranrate['Rating']
    #Applies a sum to the toptransactions after grouping it up by itemcode
    temptopTransactionrating = toptranrate.groupby('ItemCode').sum()[['similarityIndex','weightedrating']]
    temptopTransactionrating.columns = ['sum_similarityIndex','sum_weightedRating']
    
    #Creates an empty dataframe
    recommendation_df = pd.DataFrame()
    #Now we take the weighted average
    recommendation_df['weighted average recommendation score'] = temptopTransactionrating['sum_weightedRating']/temptopTransactionrating['sum_similarityIndex']
    recommendation_df['ItemCode'] = temptopTransactionrating.index
    recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
    result = recommendation_df.head(2)['ItemCode'].tolist()
    return result
```
